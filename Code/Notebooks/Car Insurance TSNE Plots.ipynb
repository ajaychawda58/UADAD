```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
```

```python
from utils import *
```

```python
data = load_data('../../data/car_insurance/insurance_claims.csv')
categorical_cols =  ['policy_state', 'umbrella_limit', 'insured_sex', 'insured_education_level',
    	'insured_occupation', 'insured_hobbies', 'insured_relationship', 'incident_type',
        'collision_type', 'incident_severity', 'authorities_contacted', 'incident_state', 'incident_city',	
        'property_damage', 'police_report_available', 'auto_make', 'auto_model']
```

```python
cols_to_remove =  ['policy_number', 'policy_bind_date', 'policy_csl', 'incident_location', 'incident_date', '_c39']
X = remove_cols(data,cols_to_remove)
Y = get_labels(data, "vehicle_claims")
```

```python
X_num = X
Y = Y
X_num = remove_cols(X_num, categorical_cols)
```

**Numerical Cols**

```python
X_num = normalize_cols(X_num)
X_num_embedded = TSNE(n_components=2, learning_rate='auto', init='random').fit_transform(X_num)
```

```python
X_num["tsne_one"] = X_num_embedded[:,0]
X_num["tsne_two"] = X_num_embedded[:,1]
X_num["y"] = Y
```

```python
 sns.scatterplot(
    x= "tsne_one", y= "tsne_two",
    hue="y",
    data=X_num,
    legend="full",
    alpha=1
)
```

**Label Encoding**

```python
X_label = label_encoding(X, categorical_cols)
X_label = normalize_cols(X_label)
X_label = X_label
```

```python
X_label_embedded = TSNE(n_components=2, learning_rate='auto', init='random').fit_transform(X_label)
```

```python
X_label["tsne_one"] = X_label_embedded[:,0]
X_label["tsne_two"] = X_label_embedded[:,1]
X_label["y"] = Y
```

```python
sns.scatterplot(
    x= "tsne_one", y= "tsne_two",
    hue="y",
    data=X_label,
    legend="full",
    alpha=1
)
```

**GEL Encoding**

```python
X_gel = gel_encoding(X, categorical_cols)
X_gel = normalize_cols(X_gel)
```

```python
X_gel_embedded = TSNE(n_components=2, learning_rate='auto', init='random').fit_transform(X_gel)
```

```python
X_gel["tsne_one"] = X_gel_embedded[:,0]
X_gel["tsne_two"] = X_gel_embedded[:,1]
X_gel["y"] = Y
```

```python
sns.scatterplot(
    x= "tsne_one", y= "tsne_two",
    hue="y",
    data=X_gel,
    legend="full",
    alpha=1
)
```

```python
X_oh = X
X_oh = one_hot_encoding(X_oh, categorical_cols)
X_oh = normalize_cols(X_oh)
```

```python
X_oh_embedded = TSNE(n_components=2, learning_rate='auto', init='random').fit_transform(X_oh)
```

```python
X_oh["tsne_one"] = X_oh_embedded[:,0]
X_oh["tsne_two"] = X_oh_embedded[:,1]
X_oh["y"] = Y
```

```python
 sns.scatterplot(
    x= "tsne_one", y= "tsne_two",
    hue="y",
    data=X_oh,
    legend="full",
    alpha=1
)
```

```python
f, ax = plt.subplots(2,2, figsize = [10,10])
sns.scatterplot(
    x= "tsne_one", y= "tsne_two",
    hue="y",
    data=X_num,
    legend="full",
    alpha=1, ax= ax[0,0]
)
sns.scatterplot(
    x= "tsne_one", y= "tsne_two",
    hue="y",
    data=X_label,
    legend="full",
    alpha=1, ax=ax[0,1]
)
sns.scatterplot(
    x= "tsne_one", y= "tsne_two",
    hue="y",
    data=X_gel,
    legend="full",
    alpha=1, ax=ax[1,0]
)
sns.scatterplot(
    x= "tsne_one", y= "tsne_two",
    hue="y",
    data=X_oh,
    legend="full",
    alpha=1, ax=ax[1,1]
)
ax[0][0].set_title("Numerical Columns")
ax[0][1].set_title("Label Encoding")
ax[1][0].set_title("Gel Encoding")
ax[1][1].set_title("One hot Encoding")
```
