**Add repair hous and repair cost based on complexity**
\
*Every 10th row has anomalous price and every 20th row has anomalous working hours*

```python
def add_hours_and_cost(data):
    repair_hours = []
    repair_cost = []
    label = []
    issue = data['issue']
    issue_id = data['issue_id']
    price = data['Price']
    complexity = data['repair_complexity']
    count = 0
    for i in range(len(data)):
        count += 1
        if issue[i] == 'Steering Wheel Shaking':
            hour = complexity[i]*1
            cost = hour * 20 + price[i]*0.001
        elif issue[i] == 'Brake Pads Worn':
            hour = complexity[i]*2
            cost = hour * 20 + price[i]*0.0005
        elif issue[i] == 'Flat Tyres':
            hour = complexity[i]*1
            cost = hour * 20 + price[i]*0.0003
        elif issue[i] == 'Tyre Alignment':
            hour = complexity[i]*0.5
            cost = hour * 20 + price[i]*0.0001
        elif issue[i] == 'Alternator Failing':
            hour = complexity[i]*2
            cost = hour * 20 + price[i]*0.02
        elif issue[i] == 'Radiator Leaking':
            hour = complexity[i]*2
            cost = hour * 20 + price[i]*0.02
        elif issue[i] == 'Starter Motor Issue':
            hour = complexity[i]*3
            cost = hour * 20 + price[i]*0.01
        elif issue[i] == 'Windscreen Crack':
            hour = complexity[i]*1
            cost = hour * 20 + price[i]*0.0006
        elif issue[i] == 'Excessive Emissions':
            hour = complexity[i]*1
            cost = hour * 20 + price[i]*0.0009
        elif issue[i] == 'Sensor Malfunction':
            hour = complexity[i]*3
            cost = hour * 20 + price[i]*0.05
        elif issue[i] == 'Gear Box Issue':
            hour = complexity[i]*2
            cost = hour * 20 + price[i]*0.01
        elif issue[i] == 'Warning Light':
            if issue_id[i] == 1:
                hour = complexity[i]*1
                cost = hour * 20 + price[i]*0.001
            elif issue_id[i] == 2:
                hour = complexity[i]*0.5
                cost = hour * 20 + price[i]*0.005
            elif issue_id[i] == 3:
                hour = complexity[i]*2
                cost = hour * 20 + price[i]*0.003
            elif issue_id[i] == 4:
                hour = complexity[i]*5
                cost = hour * 20 + price[i]*0.005
            elif issue_id[i] == 5:
                hour = complexity[i]*3
                cost = hour * 20 + price[i]*0.004
            elif issue_id[i] == 6:
                hour = complexity[i]*2
                cost = hour * 20 + price[i]*0.002
            elif issue_id[i] == 7:
                hour = complexity[i]*1
                cost = hour * 20 + price[i]*0.004
            else:
                hour = complexity[i]*9
                cost = hour * 20 + price[i]*0.01
        elif issue[i] == 'Electrical Issue':
            if issue_id[i] == 1:
                hour = complexity[i]*2
                cost = hour * 20 + price[i]*0.001
            elif issue_id[i] == 2:
                hour = complexity[i]*0.5
                cost = hour * 20 + price[i]*0.002
            elif issue_id[i] == 3:
                hour = complexity[i]*1
                cost = hour * 20 + price[i]*0.005
            elif issue_id[i] == 4:
                hour = complexity[i]*2
                cost = hour * 20 + price[i]*0.003
            else:
                hour = complexity[i]*3
                cost = hour * 20 + price[i]*0.001
        elif issue[i] == 'Engine Issue':
            if issue_id[i] == 1:
                hour = complexity[i]*8
                cost = hour * 20 + price[i]*0.2
            elif issue_id[i] == 2:
                hour = complexity[i]*16
                cost = hour * 20 + price[i]*0.15
            elif issue_id[i] == 3:
                hour = complexity[i]*12
                cost = hour * 20 + price[i]*0.1
            else:
                hour = complexity[i]*10
                cost = hour * 20 + price[i]*0.05
        else:
            if issue_id[i] == 1:
                hour = complexity[i]*4
                cost = hour * 20 + price[i]*0.003
            elif issue_id[i] == 2:
                hour = complexity[i]*3
                cost = hour * 20 + price[i]*0.007
            else:
                hour = complexity[i]*8
                cost = hour * 20 + price[i]*0.009
        if count % 10 == 0:
            repair_cost.append(cost*random.uniform(3,6))
            label.append(1)
        else:
            repair_cost.append(cost)
            label.append(0)
        if count % 20 == 0:
            repair_hours.append(cost*random.uniform(3,4))
            
        else:            
            repair_hours.append(hour)
            
    df = pd.DataFrame(np.array([repair_cost]).T)
    df.columns = ['repair_cost']
    df1 = pd.DataFrame(np.array([repair_hours]).T)
    df1.columns = ['repair_hours']
    df2 = pd.DataFrame(np.array([label]).T)
    df2.columns = ['Label']
    data = pd.concat([data, df, df1], axis=1)
    data = pd.concat([data, df2], axis=1)
    return data
        
```
